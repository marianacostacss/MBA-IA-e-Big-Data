### **Questão 01**.
#Carregue o dataset a partir do arquivo fornecido. Depois conte quantos valores ausentes existem para cada atributo.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)
```

---
### **Questão 02.**
#Separe o conjunto em $80\%$ treino e $20\%$ teste.

stratify=y --> Mantém a proporção

random_state=1234 --> Define e mantém a mesma semete geradora


```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=1234)
```

```python
p = len(X_train)
q = len(X_test)
print(p)
print(q)
```

---
### **Questão 03**.
#Treine um Classificador usando o método de Árvore de Decisão  no conjunto de treino e extraia a acurácia, precisão e recall no conjunto de teste.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score

dtc = DecisionTreeClassifier(random_state=42) #segue o mesmo padrão e numero de amostras


dtc.fit(X_train, y_train) #ajustanto árvore ao conjunto de treinamento


y_pred = dtc.predict(X_test) #predição

print("%.2f" % accuracy_score(y_test, y_pred)) #comparação de teste e predição
print("%.2f" % precision_score(y_test, y_pred, average='micro'))
print("%.2f" % recall_score(y_test, y_pred, average='micro'))
```

### **Questão 04**.

#Visualize a árvore de decisão resultante (`sklearn.tree.plot_tree`) do treinamento e diga quais são a feature mais informativa. Justifique.

```python
from sklearn.tree import plot_tree #visualização da árvore
import matplotlib.pyplot as plt
plt.figure(figsize=(12,12))
plot_tree(dtc, filled=True, max_depth=None)
```

Imprimir um relatorio das classes separadas.

```python
target_names = ['iris-setosa', 'iris-virginica', 'iris-versicolor']
print(classification_report(y_test, y_pred, target_names=target_names))
```
